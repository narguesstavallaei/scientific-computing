$\color{#3cb371}{A = LU}$

```python
import numpy as np
```

<span style="color:#3cb371">Without Pivoting</span>

```python
def LU(A):
    m = A.shape[0]
    U = A.copy()
    L = np.eye(m)
    for k in range(m-1):
        for j in range(k+1,m):
            L[j,k] = U[j,k] / U[k,k]
            U[j,k:] = U[j,k:] - L[j,k] * U[k,k:]
    return L, U
```

<span style="color:#3cb371">Enter an arbitrary square matrix A and find its LU decomposition.</span>

```python
A=np.array([[2,1],[4,3]], dtype=float)
L, U = LU(A)
print("\033[91m L =", "\033[0m")
for row in L:
    print("\033[91m" + str(row) + "\033[0m")
print("\033[91m U =", "\033[0m")
for row in U:
    print("\033[91m" + str(row) + "\033[0m")
```

<span style="color:#3cb371">Find </span> $\color{#3cb371}{\|A-LU\|_{\infty}}$


```python
print("\033[91m max|A-LU| =  ", np.abs(A - L @ U).max(), "\033[0m")
```

<span style="color:#3cb371"> Solution using LU decomposition </span> 
$$\color{#3cb371}{LUx=b}$$
<span style="color:#3cb371"> **Step 1**: solve </span> 
 $\color{#3cb371}{Ly = b}$ <span style="color:#3cb371"> using forward substitution </span> 
 
<span style="color:#3cb371"> **Step 2**: solve </span> 
 $\color{#3cb371}{Ux=y}$ <span style="color:#3cb371"> using backward substitution </span> 

```python
def LUSolve(L,U,b):
    m = L.shape[0]
    # solve Ly = b
    y = np.empty_like(b)
    for i in range(m):
        y[i] = b[i] - np.dot(L[i,0:i], y[0:i])
    # solve Ux = y
    x = np.empty_like(b)
    for i in range(m-1,-1,-1):
        x[i] = (y[i] - np.dot(U[i,i+1:m],x[i+1:m])) / U[i,i]
    return x
```

<span style="color:#3cb371"> Select vector b, find the solution of </span> 
$\color{#3cb371}{Ax=b,}$ <span style="color:#3cb371"> and compute the error using the infinity norm. </span>

```python
b = np.array([2.01,1], dtype=float)
x = LUSolve(L,U,b)
print("\033[91m x =", x, "\033[0m")
print("\033[91m ||Ax-b||_sup = ", np.linalg.norm(A@x - b, np.inf), "\033[0m")
```
